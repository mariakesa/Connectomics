```python
import networkx as nx
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fname='/home/maria/Downloads/drosophila_medulla_1.graphml'
graph=nx.read_graphml(fname)
```

```python
def parse_graph(graph):
    adj_lst={}
    for line in nx.generate_adjlist(graph):
        strp_n=line.replace("n", "")
        lst=strp_n.split(' ')
        int_lst=[]
        for ind in lst:
            int_lst.append(int(ind))
        adj_lst[int_lst[0]]=int_lst[1:]
    return adj_lst
```

```python
adj_lst=parse_graph(graph)
```

```python
def adj_lst_to_mat(adj_lst):
    n_nodes=len(adj_lst.keys())
    con_mat=np.zeros((n_nodes,n_nodes))
    for node in adj_lst.keys():
        for adj in adj_lst[node]:
            con_mat[node,adj]=1
    return con_mat
```

```python
adj_mat=adj_lst_to_mat(adj_lst)
```

```python
sns.heatmap(adj_mat)
plt.show()
```

```python
def histogram_of_neighbors(adj_lst):
    adj_len=[]
    for node in adj_lst.keys():
        adj_len.append(len(adj_lst[node]))
    return adj_len
```

```python
adj_len=histogram_of_neighbors(adj_lst)
plt.hist(adj_len,bins=100)
```
